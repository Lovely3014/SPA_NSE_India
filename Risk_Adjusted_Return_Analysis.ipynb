# Risk Adjusted Return Analysis

```python
df['Annual_returns'] = df['Close'].mean() * 252
```

```python
df['Annual_volatility'] = df['Close'].std() * (252**0.5)
```

```python
rfr = 0.01
```

```python
df['Sharp_ratio'] = df.Annual_returns - rfr / df.Annual_volatility
df
```
